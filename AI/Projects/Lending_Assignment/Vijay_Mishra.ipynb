```python
#Univariate Analysis - Loan Amoount verus loan status
#find the total number of borrowers according to loan status and load category
LoanDataWithDropedCol1=LoanDataWithDropedCol[['loan_status','loan_amnt']]
LoanDataWithDropedCol1['loan_cat']=LoanDataWithDropedCol['loan_amnt'].apply(lambda x:"Low" if x<=10000 else "Medium" if 10000<x<=20000 else "High")
LoanDataWithDropedCol2=LoanDataWithDropedCol1.groupby(['loan_cat','loan_status']).size().reset_index(name="Totals")
LoanDataWithDropedCol2
```

```python
#show bar chart total borrowers according to loan category
import matplotlib.pyplot as plt
import numpy as np
 

raw_data = {'Loan Category': ['High', 'Medium', 'Low'],
        'Charged Off': [865, 1806, 2956],
        'Fully Paid': [3406, 19318, 10226]}
            
df = pd.DataFrame(raw_data, columns = ['Loan Category', 'Charged Off', 'Fully Paid'])

# Setting the positions and width for the bars
pos = list(range(len(df['Charged Off']))) 
width = 0.25 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(15,5))

# in position pos,
plt.bar(pos, 
        df['Charged Off'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in first_name
        label=df['Loan Category'][0]) 

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        df['Fully Paid'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in first_name
        label=df['Loan Category'][1]) 


# Set the y axis label
ax.set_ylabel('Total number of Borrowers')


ax.set_xlabel('Loan Category')

# Set the chart's title
ax.set_title('Total Defaulter A/C to Category')

# Set the position of the x ticks
ax.set_xticks([p + 0.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(df['Loan Category'])

# Adding the legend and showing the plot
plt.legend(['Charged Off', 'Fully Paid'], loc='upper left')
plt.grid()
plt.show()
```
