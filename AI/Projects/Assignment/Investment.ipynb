```python
#checkpoint6.1


import matplotlib.pyplot as plt
import numpy as np
 

raw_data = {'funding_type': ['venture', 'seed', 'private equity'],
        'Total Investment': [59012.62e+07, 1698.824e+07, 14192.54e+07],
        'Average Investment': [1.063425e+07, .05566629e+07, 6.211179e+07]}
            
df = pd.DataFrame(raw_data, columns = ['funding_type', 'Total Investment', 'Average Investment'])

# Setting the positions and width for the bars
pos = list(range(len(df['Total Investment']))) 
width = 0.25 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(15,5))

# in position pos,
plt.bar(pos, 
        df['Total Investment'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in first_name
        label=df['funding_type'][0]) 

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        df['Average Investment'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in first_name
        label=df['funding_type'][1]) 


# Set the y axis label
ax.set_ylabel('Total Investment Amount')


ax.set_xlabel('Funding Type')

# Set the chart's title
ax.set_title('Investment Choise')

# Set the position of the x ticks
ax.set_xticks([p + 0.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(df['funding_type'])

# Setting the x-axis and y-axis limits
#plt.xlim(min(pos)-width, max(pos)+width*4)
#plt.ylim([0, max(df['Total Investment'] + df['Average Investment'])] )

#ax.set_ylim(100000,100000000)
#plt.ylim([100000, max(df['Total Investment'] + df['Average Investment'])] )
plt.ylim([100000, 100000000] )

# Adding the legend and showing the plot
plt.legend(['Total Investment', 'Average Investment'], loc='upper left')
plt.grid()
plt.show()
```
